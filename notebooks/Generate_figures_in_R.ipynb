## 1. Generate categories chord diagram

Install the circlize library if we haven't done this already.

```r
install.packages("circlize", repos='http://cran.us.r-project.org')
```

Load dependencies and co-occurrence matrix.

```r
library(circlize)
comatrix = readRDS('../data/category_cooccurrences.rds')
comatrix
```

Format it for the chord diagram function. Looks the same, but it changes the type from data.frame to data.matrix and sets the colnames and rownames again, which apparently matters.

```r
data = data.matrix(comatrix)
colnames(data) = colnames(comatrix)
rownames(data) = colnames(comatrix)
data
```

Visualize the chord diagram. 

```r
par(cex=1.5, # scales text by 150%
    height=5)

chordDiagram(data, 
             transparency = 0.4, 
             symmetric=FALSE,       # When symmetric = TRUE, only lower triangular matrix WITHOUT diagonal will be used
             self.link=1)           # 2 styles for self linking, 1 and 2 (1 is self-contained, 2 is a loop back)
```

Alternative visualization that might be more confusing, but the colors at the bottom all divide neatly between subcategories.

```r
chordDiagram(comatrix)
```
